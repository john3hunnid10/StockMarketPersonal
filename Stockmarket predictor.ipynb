```python
#imports
from IPython.display import display, HTML
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from types import AsyncGeneratorType
import datetime as dt
import warnings
#installing/importing joblib
!pip install mlxtend
import joblib
import sys
sys.modules['sklearn.externals.joblib'] = joblib
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
#installing/import yahoofinance
!pip install yahoofinance
import yfinance as yf



#importing models
Hmodel=joblib.load('highModel.joblib')
Lmodel=joblib.load('lowModel.joblib')
Omodel=joblib.load('openModel.joblib')
Cmodel=joblib.load('closeModel.joblib')
ACmodel=joblib.load('AcloseModel.joblib')
Vmodel=joblib.load('volumeModel.joblib')


companies=['','INTEL', 'MSI', 'AMD', 'ASUS']
values=['HIGH','LOW','OPEN','CLOSE','ADJ CLOSE','VOLUME']
models={
    'HIGH':(Hmodel,'High'),
    'LOW':(Lmodel,'Low'),
    'OPEN':(Omodel,'Open'),
    'CLOSE':(Cmodel,'Close'),
    'ADJ CLOSE':(ACmodel,'Adj Close'),
    'VOLUME':(Vmodel,'Volume')
}
warnings.filterwarnings("ignore", message="X does not have valid feature names", category=UserWarning)
warnings.filterwarnings("ignore", message="X does not have valid feature names", category=FutureWarning)
print('Companies are:',companies[1:])
print('Values are:',values)
print('Date should be integers in the form: (Year, Month, Day)')
print('Example input: ((2024,5,30),"INTEL","HIGH")')



def getRealVal(predDateSTR,nextDay,companyINT,strVal):
  #given inputs from the main function, this helper accesses yahoo finances and returns the desired value
  tickers=['','INTC','2377.TW','AMD','2357.TW']
  ticker=yf.Ticker(tickers[companyINT])
  data=ticker.history(start=predDateSTR,end=nextDay)
  realVal=data[strVal].iloc[0]
  return realVal

def makePrediction(date,company,value):
  #getting index of company, and verifying input
  if(company.upper() not in companies):
    print('Invalid Company')
    return -1
  else:
    companyINT=companies.index(company.upper())

  #getting desired value and model, and verifying input
  strVal=''
  if(value.upper() not in models):
    print('Invalid Value')
    return -1
  else:
    model,strVal=models[value.upper()]

  #getting date and verifying input, which involves converting it into a datetime object
  startDate=dt.datetime(2023,1,3)
  if(len(date)!=3):
    print('Invalid Date')
    return -1
  year,month,day=date
  if(type(year)!=int or type(month)!=int or type(day)!=int or 1>month>12 or 1>day>31):
    print('Invalid Date')
    return -1
  predDate=dt.datetime(year,month,day)
  timeSince=(predDate-startDate).days
  if(timeSince<0):
    print('Invalid Date')
    return -1

  #make the prediction
  prediction=model.predict([[timeSince,companyINT]])

  #if the given date is in the future, it doesn't access yahoo finance and returns the prediction
  if((predDate-dt.datetime.now()).days>0):
    print('Prediction is:',prediction[0])
    return prediction[0]

  #if the given date is in the past, it gets converted into a string and input to a helper function that accesses yahoo finance
  #it then gives an accuracy based on the delta of the real and predicted value
  nextDay=predDate+dt.timedelta(days=1)
  nextDay=nextDay.strftime('%Y-%m-%d')
  predDateSTR=predDate.strftime('%Y-%m-%d')

  #call to helper function
  realVal=getRealVal(predDateSTR,nextDay,companyINT,strVal)
  delta=abs(realVal-prediction[0])
  Accuracy=(1-(delta/realVal))*100

  #printing the prediction, real value, and accuracy
  print('Prediction is:',prediction[0])
  print('Real Value is:',realVal)
  print('Accuracy is:',Accuracy)

  return prediction[0]
makePrediction((2024,6,30),"INTEL","HIGH")
```
